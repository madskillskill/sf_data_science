```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation

x = np.random.uniform(-1, 1)
y = np.random.uniform(-1, 1)
# Create a figure and axis
fig, ax = plt.subplots()

# Set the limits of the plot
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])


distance_text = ''
# Create a dot object
dot, = ax.plot([], [], 'ro')

# Create a line object
line, = ax.plot([], [], 'b-')

distance_text = ax.text(0, 1, s='', transform=ax.transAxes)

# L=plt.legend(loc=1) #Define legend objects


def update_axis(n):
    if n >= 10:
        n -= np.random.uniform(0,1)
        return (n)
    elif n <= -10:
        n += np.random.uniform(0,1)
        return (n)
    else:
        n -= np.random.uniform(-1,1)
        return(n)
        
def update_point(point):
    point_x = update_axis(point[0])
    point_y = update_axis(point[1])
    point = [point_x, point_y]
    return point




# Function to initialize the animation
def init():
    dot.set_data([], [])
    line.set_data([], [])
    return dot, line

# Function to update the animation on each frame
def update(frame):
    global x
    global y
    # Generate random x and y coordinates for the dot
    x = update_axis(x)
    y = update_axis(y)
    print('x = ',x,'::: y = ', y)
    
    # Update the dot and line data
    dot.set_data(x, y)
    line.set_data([0, x], [0, y])   
     
    # Calculate the distance from the center to the dot
    distance = np.sqrt(x**2 + y**2)
    
    distance_text.set_text('Distance from center: %.3f J' % distance)
    # distance_text.set_text(s=["Distance from center:", distance])
    # global distance_text
    # distance_text = distance
    

    
    # Print the dot's distance from the center
    # print("Distance from center:", distance)
    
    return dot, line







# Create the animation
ani = animation.FuncAnimation(fig, update, frames=100,  interval = 100, blit=True)

# Show the plot
plt.show()
print(distance)
```
