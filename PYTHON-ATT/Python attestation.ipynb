```python
import pandas as pd # Импорт нужных библотек
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
sd_df = pd.read_csv('data/solarpower.csv')
print(sd_df.head())
print(sd_df['cum_power'].describe())
```

```python

for i,row in sd_df.iterrows():
    if row['cum_power'] == -1:
        sd_df.loc[i, "cum_power"]  =  (sd_df.loc[i-1, "cum_power"] + sd_df.loc[i+1, "cum_power"])/2

print(sd_df['cum_power'].describe())
```

```python
import datetime as dt
sd_df['day_power'] = sd_df['cum_power']
print(sd_df.info())
sd_df['date'] = pd.to_datetime(sd_df['date'])
print(sd_df.info())
```

```python
mask = sd_df[(sd_df['date'] >= '2017-02-28') & (sd_df['date'] <= '2017-04-01')]
```

```python
print(mask['day_power'].describe())
for i,row in mask.iterrows():
    while i < len(sd_df['day_power']):
        if i == 0:
            mask.loc[i, "day_power"] = 0
        else:
            mask.loc[i, "day_power"]  =  (sd_df.loc[i, "day_power"] - sd_df.loc[i-1, "day_power"])
        i += 1

print(sd_df['day_power'].describe())

mask['day_power'].mean()
```
